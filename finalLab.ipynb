```python
def newImg(image_file):
    f = io.imread(image_file)
    grayscale = rgb2gray(f)
    rows = grayscale.shape[0]
    cols = grayscale.shape[1]
    newImageMade = np.zeros((rows,cols,3),dtype='uint8')
    for i in range(rows):
        for j in range(cols):
            if grayscale[i,j] > 0.75:
                newImageMade[i,j] = [1,0,0.5]
            elif grayscale[i,j] > 0.5:
                newImageMade[i,j] = [0.4,0.5,0]
            elif grayscale[i,j] > 0.25:
                newImageMade[i,j] = [0,0,1]
            else:
                newImageMade[i,j] = [1,1,1]
            
        return plt.imshow(newImageMade)
```

```python
newImg('james_profile.jpg')
```
